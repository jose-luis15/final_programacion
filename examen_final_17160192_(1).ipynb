# 2.2 Hacer una consulta al DataFrame para extraer los datos de Incendios que pertenezcan a los departamentos de Loreto en el año 2010.


para esta pregunta pude corroborar durante el examen que el archivo no tenia datos de loreto, asi que para el presente solucionario tomare a PUNO para el desarrollo de esta pregunta y las que siguen, puesto que este departamento presenta casos de incendio, que ocurrieron en el año 2010

```
# incendios en PUNO ocurridos en el 2010
lor = incendios[(incendios["NOMBPROV"] == "PUNO") & (incendios["INCENDIO"] == "SI") & (incendios["ANNO"] == 2010)]
lor
```

# 2.3 Usando numpy o pandas ¿cuántas provincias de Puno han sido afectadas en el 2010?

```
# para ver la cantidad de provincias afectadas en el 2010 usamos:
len(lor)
```

extraemos las coordenas este y norte del campo coordenasa de nuestro DataFrame, para lograr esto se crearan dos listas vacias (lista1 y lista2) a las cuales luego se le agregaran las coordenadas este y norte extraidas de la columna coordenadas que seran separadas haciendo uso de la herramienta "split", luego estas dos listas seran añadidar a nuestra data en forma de columna indicando el este y norte de los puntos analizados

```
lista1 =[]
lista2 =[]

for x in lor["COORDENADAS"]:
  b = x.split(",")
  lista1.append(float(b[0]))
  lista2.append(float(b[1]))
  
lor["ESTE"] = lista1
lor["NORTE"] = lista2
lor.drop(['COORDENADAS'], axis=1)

```

# 2.4 Generar un GeoDataFrame para la consulta a partir de las coordenadas.


para esta pregunta he desarrollado un codigo para consultar las direcciones de cada coordenada que se extrajo de las zonas de Puno donde ocurrieron incendios durante el 2010

```
from geopy.geocoders import Nominatim
```

```
nominacion = Nominatim()
w = []
for x in list(zip(lista2,lista1)):
  w.append(str(x))
for z in w:
  puntos = nominacion.reverse(z)
  print(puntos.address)
```

# 2.5 Exportar la data creada a un archivo shapefile y visualizar los datos usando Folium.


```
# llamareamos a tres librerias con las que trabajaremos(fiona, geopandas y shapely)
# en caso de que no esten instalados se les podra insertar anteponiendo "!pip install" al nombre de la libreria
!pip install fiona
!pip install geopandas
from geopandas import GeoDataFrame
from shapely.geometry import Point
import fiona
```

```
# para poder trabajar con folium necesitamos que nuestra data sea un geodatframe antes de un json, pero 
# esto se lograra si nuestro archivo tiene el campo geometria, el cual le dara la "espacialidad"
# asi mismo elegiremos el sistema de coordenadas para nuestros puntos
geometry = [Point(xy) for xy in zip(lor.ESTE, lor.NORTE)]
crs = {'init' : 'epsg:4326'}
geo_df = GeoDataFrame(lor, crs=crs , geometry = geometry)
```

```
geo_df.plot()
```

```
# convertimos nuestro geojson a json para abrirlo con la libreria folium
gjson = geo_df.to_json()
```

# 2.6 Agregar el departamento que te corresponde al mapa folium. (leer archivo departamentos.json) *Extra

```
# leemos el archivo json de la siguiente manera, si miramos la tabla que se imprimio en pantala 
# notaremos que la columna geometria indica que se trata de un archivo en forma de poligono
data = GeoDataFrame.from_file("departamentos.json")
data.head()
```

```
# como nuestra zoa de interes es Puno extraeremos el poligono que le corresponde a este departamento
puno = data[data["DEPNOM"] == "PUNO"]
```

```
puno.plot()
```

```
# para poder exportar el poligono de Puno dentro de folium sera necesario que este sea transformado a un json, pues el anterior archivo(leido)
# esta en formado GeoDataFrame
puno = data[data["DEPNOM"] == "PUNO"].to_json()
```

ahora procedemos a exportar el poligono y los puntos sobre el mismo tile generado a partir de la libreria folium

```
import folium
mapa = folium.Map([-16,-70], zoom_start=7, tiles="openstreetmap") #condiciones del mapa
poligono = folium.features.GeoJson(puno)
poligono.add_to(mapa)
puntos = folium.features.GeoJson(gjson) #importamos los puntos(que estan en formato geojson)
mapa.add_children(puntos) #añadimos los puntos al mapa creado
mapa
```
