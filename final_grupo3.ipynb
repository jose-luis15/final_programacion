```
import pandas as pd
```

```
a = pd.read_excel("bici_sanisidro.xlsx", sep =";")
a.head()
```

```
lista1 =[]
lista2 =[]

for x in a["COORDENADAS"]:
  b=x.split(";")
  lista1.append(float(b[0]))
  lista2.append(float(b[1]))
print(lista2)
print(lista1)
```

```
a["ESTE"] = lista1
a["NORTE"] = lista2
a.drop(['COORDENADAS'], axis=1)
```
